```python
#Imputing the null values of Response Columns
```

```python
cs_data['AcceptedAnyCampaign'] = cs_data[['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']].max(axis=1)
cs_data['AcceptedAnyCampaign'] 
```

```python
cs_data['Response'] = cs_data['Response'].fillna(cs_data['AcceptedAnyCampaign'])
```

```python
variables = ["Income", "MntWines", "MntMeatProducts", "MntFishProducts", "MntSweetProducts","MntSnacksProds", "MntFruits" ]
cs_data_subset = cs_data[variables]

fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(12, 7))

for i, ax in enumerate(axs.flatten()):
    if i < len(variables):
        variable = variables[i]
        ax.boxplot(cs_data_subset[variable])
        ax.set_title(variable)

plt.tight_layout()
plt.show()
```
