```python
#Creating New Feature Based On Existing columns
```

```python
cs_data['DateTime'] = pd.to_datetime(cs_data['Year_Birth'])
```

```python
cs_data['Year_Birth']
```

```python
cs_data['Age'] = 2023 - cs_data['Year_Birth']
cs_data['TotalChildren'] = cs_data['Kidhome'] + cs_data['Teenhome']
cs_data['MonthsSinceCustomer'] = (pd.to_datetime('today') - pd.to_datetime(cs_data['Dt_Customer'])).dt.days / 30
cs_data['TotalSpent'] = cs_data['MntWines'] + cs_data['MntFruits'] + cs_data['MntMeatProducts'] + cs_data['MntFishProducts'] + cs_data['MntSweetProducts'] + cs_data['MntSnacksProds']
cs_data['NumTotalPurchases'] = cs_data['NumDealsPurchases'] + cs_data['NumWebPurchases'] + cs_data['NumCatalogPurchases'] + cs_data['NumStorePurchases']
cs_data['PurchasesPerVisit'] = cs_data['NumTotalPurchases'] / cs_data['NumWebVisitsMonth']
cs_data['HasComplained'] = cs_data['Complain'].apply(lambda x: 1 if x == 1 else 0)
cs_data['HasResponded'] = cs_data['Response'].apply(lambda x: 1 if x == 1 else 0)
cs_data['HasAccepted'] = cs_data['AcceptedAnyCampaign'].apply(lambda x: 1 if x == 1 else 0)
```
